```python
#Yeni bir kolon ekleyelim usd-try degisim oranını yüzde bazlı ekleyelim
data['degisim(usd/try)']=data['usd_try'].pct_change()*100

print(data.head())


#Görselleştirme denemesi

import matplotlib.pyplot as plt


# "Tarih" sütununu datetime formatına çevirelim
data['Tarih'] = pd.to_datetime(data['Tarih'], format='%Y-%m')

# Grafiği yeniden oluşturalım
plt.figure(figsize=(10,6))
plt.plot(data['Tarih'], data['degisim(usd/try)'], marker='o', linestyle='-', color='b')
plt.title('USD/TRY Değişim Oranı (%)')
plt.xlabel('Tarih')
plt.ylabel('Değişim Oranı (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
